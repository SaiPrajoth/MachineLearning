```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv('/Users/saiprajothmallyala/Downloads/canada_per_capita_income.csv')
data_year = data.drop('pci',axis='columns')
```

```python
new_data=data.drop('year',axis='columns')
new_data
```

```python
model = linear_model.LinearRegression();
model.fit(data_year,new_data.pci)
```

```python
model.predict([[1943],[2024],[2025]])
```

```python
p=model.predict(data_year)
```

```python
data["predicted values of pci"]=p
data
```

```python
data.to_csv('pci_prediction.csv')
```
