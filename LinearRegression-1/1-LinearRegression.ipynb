```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

# importing the libraries numpy, pandas, matplot.lib and sklearn
```

```python
df = pd.read_csv('/Users/saiprajothmallyala/Downloads/house-prices.csv')
new_df=df.drop('Price',axis='columns')
new_df

# reading the csv file which contains the data to be used to train the model
# Traditinallly before doing any of the process, we do data cleaning and data processing to avoid any errors in the data, like NA or Null values and others.
```

```python
# Before training any model or creating model, visualization of the data is important to get the notion of the data and understand which model will be most efficient to work on the data.


x = df.Price # Random x-coordinates
y = df.SqFt# Random y-coordinates

# Create a scatter plot
plt.scatter(x, y, label='Scatter Plot', color='blue', marker='o')

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Scatter Plot for home and prices')
```

```python
# Creation and training the model 
model = linear_model.LinearRegression()
model.fit(new_df,df.Price)
```

```python
# Testing and data prediction
model.predict([[3300],[9000]])
```

```python
# Here in the Linear Regression model, we try to find a line that has the least mean squared error (MSE) by changing the slope(s) (m) and intercept (c). The linear_model module's Linear Regression funciton will give the line equation which we need. By adding the values of the input data to it we can predict the output values
# We calculate the coefficient(s)/slopes of the model (here it is a line)

model.coef_
```

```python
# We calculate the intercept of the model (here it is a line)
model.intercept_
```

```python
# Saving the predicted values in the input csv file
p=model.predict(new_df
             )
```

```python
new_df['Predicted Prices']=p
```
